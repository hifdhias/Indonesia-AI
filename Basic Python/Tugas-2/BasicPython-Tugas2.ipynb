```python
Kontak={"Nama":[], "No Telepon":[]}

print("Selamat datang!\n")
while True:
    print("\n--- Menu ---\n1. Daftar Kontak\n2. Tambah Kontak\n3. Keluar\n")
    Menu=input("Pilih Menu: ")
    if Menu=='1':
        print("Daftar Kontak:")
        for i in range(len(Kontak["Nama"])):
                print ("Nama: " + Kontak["Nama"][i])
                print ("No Telepon: " + Kontak["No Telepon"][i])
    elif Menu=='2':
        print("Tambah Kontak: ")
        Kontak["Nama"].append(input("Nama: "))
        Kontak["No Telepon"].append(input("No Telepon: "))
        print("Kontak berhasil ditambahkan ")
    elif Menu=='3':
        print("Program selesai, sampai jumpa! ")
        break
    else:
        print("Menu tidak tersedia ")
    
```
